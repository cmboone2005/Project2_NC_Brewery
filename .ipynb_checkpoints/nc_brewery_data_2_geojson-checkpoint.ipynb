```python
beer_data_new = beer_data[["state", "city", "name_breweries", "name", "style_name", "cat_name", "geometry"]]
beer_data_new.head()
```

```python
nc_beer_geojson_df = updated_beer_data.loc[updated_beer_data["State"] == "North Carolina", :]
nc_beer_geojson_df.head(10)
```
